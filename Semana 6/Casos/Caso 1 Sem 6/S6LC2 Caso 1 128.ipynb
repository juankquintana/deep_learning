```python
# Funcin para graficar la historia del entrenamiento (preparada en talleres anteriores)
import matplotlib.pyplot as plt

def plot_history(history):
    """
    Plots training and validation loss and accuracy.

    Parameters:
      history: A dictionary containing keys:
          'loss', 'val_loss', 'accuracy', and 'val_accuracy'.
          (If a Keras History object is passed, its `.history` attribute will be used.)
    """
    hist = history.history if hasattr(history, 'history') else history

    # Final metrics
    print(f"Train Loss final: {hist['loss'][-1]:.3f}, "
          f"Train Accuracy final: {hist['accuracy'][-1]:.3f}")
    print(f"Validation Loss final: {hist['val_loss'][-1]:.3f}, "
          f"Validation Accuracy final: {hist['val_accuracy'][-1]:.3f}")

    plt.figure(figsize=(12, 5))

    # Loss plot
    plt.subplot(1, 2, 1)
    plt.plot(hist['loss'], label='Training Loss', marker='o')
    plt.plot(hist['val_loss'], label='Validation Loss', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training vs. Validation Loss')
    plt.legend()
    plt.grid(True)

    # AUC plot
    plt.subplot(1, 2, 2)
    plt.plot(hist['accuracy'], label='Training Accuracy', marker='o')
    plt.plot(hist['val_accuracy'], label='Validation Accuracy', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training vs. Validation Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

### A. Uso de _CNN_ pre-entrenada para Aprendizaje por Transferencia

La estructura de modelo secuencial tipo CNN puede ser importada directamente sin necesidad de definir cada una de las capas, como conv2D, MaxPooling, etc. Muchas redes pueden ser implementadas automticamente desde _Keras_ (https://keras.io/api/applications/). Sin embargo, muchas veces estas redes son bastante genricas. Por ejemplo, muchas de ellas han sido entrenadas con los datos de _IMAGENET_ para clasificacin multiclase, pero con alrededor de 1000 clases diferentes. Cuando requerimos un uso ms especfico, por ejemplo, reconocer la marca de un vehculo o el tipo de ambiente en una fotografa, este tipo de clasificacin genrica no nos funciona muy bien.

La opcin ms directa, es hacer nuestra propia red convolucional desde cero con nuestros propios datos. Para esto, podemos tomas como inspiracin muchas de las arquitecturas que ya han sido probadas (AlexNet, VGG16, GoogleNEt, ResNet, etc.). Sin embargo, es posible que, dado que nuestros datos corresponden a una aplicacin ms especializada, no tengamos a nuestra disposicin la cantidad suficiente de imgenes para entrenar nuestra red. 

Otra opcin ms til, sera usar como base de nuestro modelo, una red ya pre-entrenada. Por ejemplo, las primeras capas de una red tipo ResNet. El principio de este enfoque, es que las primeras capas todava no se han especilizado completamente y son bastante genricas en trminos de reconocer patrones bsicos que pueden funcionar en diversos problemas. Para estas primeras capas tomamos los pesos (estimacin de los parmetros) ya fijos (recordemos que tenemos la opcin de definir el parmetro "weights"). Sobre estas capas, podemos adicionar unas nuevas capas (generalmente son del tipo completamente conectadas) y entrenar los pesos de estas con los datos que tenemos a disposicin para nuestro problema. 

Dentro de la calibracin de los modelos de _Transfer Learning_, aparecen entonces nuevos hiper-parmetros que son necesarios de calibrar, como el nmero de capas del modelo base ya preentrenado y el nmero de capas entrenables adicionales. Esto depende, por supuesto, de la cantidad de imgenes que tengamos disponibles y de la complejidad de la tarea. 

#### _Transfer Learning_ en Keras usando VGG16

La forma ms sencilla de usar _transfer learning_ con redes pre-entrenadas, es cargando la red con los pesos fijos, como por ejemplo en el modelo VGG16, usando  `weights='imagenet'`, que incluye los valores de todos los parmetros ya entrenados con los datos descritos del ILSVRC. Sin embargo, en este caso el parmetro `include_top`, que hace referencia a si se incluyen las ltimas 3 capas de tipo completamente conectadas (_fully connected_) se define como `False`, dado que se reemplazarn por capas nuevas.

Adems de esto, es necesario aadir las capas nuevas con las cuales crearemos nuestro modelo. Una vez se tiene la configuracin del modelo, se fijan los valores de las primeras capas, haciendo para cada una de las _layers_ que no se entrenan `layer.trainable = False`.

Miremos la implementacin usando la red VGG16 contenida en _Keras Applications_:

```python
#Importar modelo entrenado VGG16 con las ltimas capas
#Puede tomar algunos segundos
vgg = VGG16(input_shape=(224,224,3), weights='imagenet', include_top=False)
vgg.summary()
```

Sobre este modelo, el primer paso (que es opcional), es usar una operacin de _pooling_ global sobre la salida de la ltima capa ("block5_pool"), para pasar a una arreglo en una dimensin de tamao [1,1,512].

```python
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D

base=GlobalAveragePooling2D()(vgg.output)
```

Despus, se pueden adicionar capas completamente conectadas ("Dense") de diferentes tamaos. En este caso, aadimos una capa de 256 nodos con activacin RELU:

```python
new=Dense(128,activation='relu')(base)
```

De la misma forma, es posible adicionar capas adicionales. 

Supongamos que el nuevo problema que queremos resolver es de reconocimiento de imgenes con 5 clases diferentes de objetos. Para esto, es necesario crear la capa de salida:

```python
out_layer=Dense(5,activation='softmax')(new)
```

El ltimo paso antes de compilar el modelo, es fijar las capas cuyos pesos queremos que se mantengan fijos, mientras que las capas restantes se dejan libres para entrenar. En este caso, todas las capas anteriores del VGG16 a las capas completamente conectadas se dejarn fijas:

```python
for layer in vgg.layers:
    layer.trainable = False
```

Ahora se puede definir el nuevo modelo que va desde los _inputs_ del modelo base (VGG) hasta la capa de salida definida:

```python
from tensorflow.keras.models import Model

model=Model(inputs=vgg.input,outputs=out_layer)
```

Y con esto, ya es posible compilar el modelo. En este caso usamos _Adam_ con un problema de clasificacin multiclase. Puede ser definido a conveniencia para calibrar el modelo en trminos de hiper-parmetros:

```python
from keras.optimizers import Adam
model.compile(optimizer= Adam(learning_rate=0.005, beta_1=0.9, beta_2=0.995), loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
```

#### Problema de reconocimiento de imgenes y datos a usar

En este caso, resolveremos un problema de identificacin de especies de flores. Se tienen 5 diferentes especies que salen en los datos: daisy (margarita), dandelion (diente de len), roses (rosas), sunflowers (girasoles), y tulips (tulipanes). La base de datos contiene en total 3670 imgenes en color  y se obtiene directamente de un repositorio usando las herramientas de Tensorflow para carga de datos:

```python
import pathlib
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
archive = tf.keras.utils.get_file(origin=dataset_url, extract=True)
data_dir = pathlib.Path(archive).with_suffix('')
```

```python
print(data_dir)
```

Miremos primero el nmero total de imgenes:

```python
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)
```

Y algunas imgenes de muestra:

Ahora definimos los datos para entrenamiento y para validacin (`train_ds` y `val_ds` respectivamente) usando un tamao de 224 por 224, para que sea compatible con VGG16:

```python
batch_size = 32
img_height = 224
img_width = 224

train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

Revisemos las dimensiones en cada uno de los conjuntos de datos:

```python
for image_batch, labels_batch in train_ds:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
```

Por ltimo, antes de usar los datos, los escalamos para que estn en el rango [0,1], y no tengamos problemas en el entrenamiento:

```python
from tensorflow.keras import layers

scale_layer = tf.keras.layers.Rescaling(1./255)

train_ds = train_ds.map(lambda x, y: (scale_layer(x), y))
val_ds = val_ds.map(lambda x, y: (scale_layer(x), y))
```

Con estos elementos ya es posible entrenar el modelo usando "model.fit". _Nota:_ El "warning" referente al uso de funciones `lambda` no representa un problema ni afecta la ejecucin del cdigo y ser eliminado en prximas versiones de Tensorflow.

```python
epochs = 10
model_train = model.fit(train_ds,epochs=epochs,verbose=1,validation_data=val_ds)
```

```python
accuracy = model_train.history['accuracy']
val_accuracy = model_train.history['val_accuracy']
loss = model_train.history['loss']
val_loss = model_train.history['val_loss']
epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

#### 1. Adaptacin de la red InceptionV3 para _transfer learning_ [30 puntos]

Ac, debe preparar la red InceptionV3 para _transfer learning_. Para ello, debe remover las capas finales (completamente conectadas) usando la opcin `top=False`.

Adems, despus de usar "global pooling" para linealizar los datos, agregar una capa adicional completamente conectada con el nmero de nodos que considere adecuado (hiper-parmetro de calibracin). Por ltimo, debe agregar la capa de salida para el problema. No olvide que los parmetros anteriores a las ltimas capas no son entrenables.

```python
from tensorflow.keras.applications import InceptionV3
incv3 = InceptionV3(input_shape=(224,224,3), weights='imagenet', include_top=False)
incv3.summary()
```

```python
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D

base_incv3=GlobalAveragePooling2D()(incv3.output)
```

```python
new_incv3_128=Dense(128,activation='relu')(base_incv3) # el enunciado dice 256 pero se implementa 128
```

```python
out_layer_incv3=Dense(5,activation='softmax')(new_incv3_128)
```

```python
for layer in incv3.layers:
    layer.trainable = False
```

```python
new_incv3_128 =Model(inputs=incv3.input,outputs=out_layer_incv3)
```

```python
new_incv3_128.compile(optimizer= Adam(learning_rate=0.005, beta_1=0.9, beta_2=0.995), loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
```

```python
epochs = 10
model_train_128 = new_incv3_128.fit(train_ds,epochs=epochs,verbose=1,validation_data=val_ds)
```

#### 2. Estimacin del modelo propuesto y calibracin de gradiente descendiente [20 puntos]
Usando el modelo propuesto, estime los parmetros asociados. No olvide calibrar el modelo con los algoritmos de estimacin correspondiente. Tambin puede considerar un capa de _dropout_ antes de la capa de salida.

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
incv3_128_tn2 = InceptionV3(input_shape=(224,224,3), weights='imagenet', include_top=False)
```

```python
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
base_incv3_128_tn2=GlobalAveragePooling2D()(incv3_128_tn2.output)
```

```python
new_incv3_128_tn2=Dense(32,activation='relu')(base_incv3_128_tn2) 
```

```python
# Aadimos una capa de dropout
from tensorflow.keras.layers import Dropout
incv3_128_tn2_dp = Dropout(0.25)(new_incv3_128_tn2)
```

```python
out_layer_incv3_128_tn2=Dense(5,activation='softmax')(incv3_128_tn2_dp)
```

```python
for layer in incv3_128_tn2.layers:
    layer.trainable = False
```

```python
from tensorflow.keras.models import Model
model_incv3_128_tn2 =Model(inputs=incv3_128_tn2.input,outputs=out_layer_incv3_128_tn2)
```

```python
from keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau

optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.995) # learning rate = 0.001 en lugar de 0.005 

lr_callback = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=2,
    verbose=1,
    min_lr=1e-6
)
```

```python
model_incv3_128_tn2.compile(optimizer= optimizer, loss='SparseCategoricalCrossentropy', metrics=['accuracy'])
```

```python
epochs = 10
model_incv3_128_tn2_train = model_incv3_128_tn2.fit(train_ds,epochs=epochs,verbose=1,validation_data=val_ds,
    callbacks=[lr_callback])
```

```python
plot_history(model_incv3_128_tn2_train)
```

```python
import pandas as pd
incv3_128_tn2_history_df = pd.DataFrame(model_incv3_128_tn2_train.history)
incv3_128_tn2_history_df.to_csv("./ModelosGuardados/incv3_128_tn2_history.csv", index=False)
model_incv3_128_tn2.save("./ModelosGuardados/incv3_128_tn2_tr_model.keras")
```
