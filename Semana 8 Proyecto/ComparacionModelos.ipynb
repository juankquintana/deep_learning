```python
import matplotlib.pyplot as plt

# Datos de BLEU Scores por modelo
bleu_scores = {
    "ResNet50": [0.363395, 0.215636, 0.140402, 0.092832],
    "VGG16": [0.501170, 0.372547, 0.294499, 0.237009],
    "InceptionV3": [0.277302, 0.143201, 0.083901, 0.053432]
}

bleu_levels = ["BLEU-1", "BLEU-2", "BLEU-3", "BLEU-4"]

# Crear la grfica
plt.figure(figsize=(10, 6))
for model, scores in bleu_scores.items():
    plt.plot(bleu_levels, scores, marker='o', label=model)

plt.title("BLEU Scores por Modelo")
plt.xlabel("N-Gram (BLEU)")
plt.ylabel("BLEU Score")
plt.ylim(0, 0.6)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
