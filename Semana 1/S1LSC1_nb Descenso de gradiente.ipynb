### 1. Definicin de funcin objetivo

Para iniciar el proceso de optimizacin, debemos definir la funcin que queremos minimizar o maximizar, es decir, la funcin objetivo, que tambin podra denominarse funcin de costo, funcin de prdida o funcin de error, segn sea el caso. Para este laboratorio buscaremos minimizar una funcin simple: *$x^2$*. 

```python
# Definir funcin objetivo convexa
def fobj(x):
    return x**2.0
```

Notemos que, como la funcin es convexa, entonces el algoritmo de gradiente descendiente buscar el mnimo.

### 2. Clculo del gradiente

La idea en el algoritmo de descenso de gradiente es que en cada iteracin se d un paso en la direccin opuesta al gradiente de la funcin objetivo evaluada en el punto dado, puesto que es la direccin en la que $f(x)$ decrece ms rpido. As, procedemos a definir el gradiente de la funcin ($\nabla f$) para usarlo en nuestra implementacin del algoritmo:

```python
# Definir gradiente de la funcin objetivo
def gradiente(x):
    return 2.0*x
```

### Funcin descenso de gradiente

Luego de definir nuestra funcin objetivo y su gradiente, debemos definir un punto o solucin inicial, una tasa de aprendizaje $\alpha$ (tambin llamada _step_ o tamao del paso), y un criterio de parada para nuestro algoritmo iterativo. Estos elementos constituyen los parmetros de nuestra funcin de descenso de gradiente, en la que partimos de la solucin inicial y actualizamos la solucin en cada iteracin de la siguiente manera:

$$ X^{new} =  X^{old} - \alpha*{\nabla f(X^{old})}$$

Lo anterior, se realiza hasta cumplir con el criterio de parada del algoritmo. En la funcin que se presenta a continuacin, se seleccion como criterio de parada que la norma del gradiente sea menor al nivel de tolerancia $(||\nabla f(X)||<\epsilon)$.

```python
# Funcin de descenso de gradiente
def descenso_gradiente(fun_obj, gradiente, sol_inicial, step, tolerancia):
    # Monitoreo de solucin y funcin objetivo evaluada
    soluciones, feval, gradeval = list(), list(), list()
    # Cargar solucin inicial
    solucion = sol_inicial
    # Definir variable booleana para validar si se cumpli el criterio de parada
    stop = False
    # Definir contador de iteraciones
    num_iter = 1
    while stop==False:
        # Evaluar gradiente
        gradiente_eval = gradiente(solucion)
        # Actualizar la solucin
        solucion = solucion - step * gradiente_eval
        # Evaluar la solucin
        solucion_eval = fun_obj(solucion)
        # Almacenar solucin e imprimir resultados de la iteracin
        soluciones.append(solucion)
        feval.append(solucion_eval)
        gradeval.append(gradiente_eval)
        print("Iteracin "+str(num_iter)," - Solucin: "+str(solucion)+". Gradiente: "+str(gradiente_eval)+". Funcin objetivo: "+str(format(solucion_eval,'.4f')))
        # Actualizar contador de iteraciones
        num_iter += 1
        # Validar si se cumpli el criterio de parada, es decir, si la norma del gradiente evaluado en la solucin actual es menor a la tolerancia dada
        if (np.linalg.norm(gradiente_eval) < tolerancia):
            stop = True
    return [soluciones, feval, gradeval]
```

### 3. Definicin de parmetros inciciales

Procedemos a probar la funcin anterior para encontrar el mnimo de $x^2$ mediante nuestro algoritmo de descenso de gradiente. Los conceptos de funcin objetivo y el gradiente ya fueron definidos previamente, por lo que solo resta definir la solucin inicial, el tamao del paso y la tolerancia. Para este caso, aunque estos tres parmetros se definieron de forma aleatoria, terminan influyendo sobre la velocidad de convergencia, tiempo de ejecucin y exactitud del algoritmo, por lo que deben ser ajustados en caso de que el algoritmo no converja a la solucin.

Usamos la funcin `descenso_gradiente`, obteniendo las soluciones y el valor de la funcin objetivo en cada iteracin:

### Ejemplo multivariado

La funcin implementada para el algortimo de descenso de gradiente tambin se puede aplicar para encontrar el ptimo de una funcin con mltiples variables. La lgica sigue siendo la misma y lo nico que cambia es que $X$ y $\nabla f(X)$ ahora seran vectores:

$$ X = \left[x_1, x_2, ..., x_p\right]$$

$$ {\nabla f(X)} = \left[\frac{\partial f(X)}{\partial x_1}, \frac{\partial f(X)}{\partial x_2}, ..., \frac{\partial f(X)}{\partial x_p}\right]$$

El criterio de parada pasa a ser entonces que la norma del gradiente sea menor a la tolerancia, lo cual est implementado en la funcin anterior al usar _np.linalg.norm( )_. En el caso univariado se calculaba el valor absoluto del gradiente, mientras que al aumentar la dimensionalidad se tiene que calcular la norma del vector $p$-dimensional. 

Procedamos entonces a solucionar el siguiente problema de optimizacin en $\mathbb{R}^3$:

$$ \min f(X) = 4x_{1}^2+\frac{27}{2}x_{2}^2 + \frac{3}{2}x_{3}^2 $$

```python
# Definir funcin objetivo - X sera un np.array
def fobj(X):
    return 4*X[0]**2 + (27/2)*X[1]**2 + (3/2)*X[2]**2
# Definir gradiente de la funcin objetivo - sera una np.array con 3 elementos (la derivada parcial con respecto a cada x)
def gradiente(X):
    return np.array((8*X[0], 27*X[1], 3*X[2]))
# Definir tolerancia
tolerancia = 0.001
# Definir tamao del paso
step = 0.05
# Definir solucin inicial
sol_inicial = np.array((-50, 0.1, 20))
# Llamar la funcin descenso de gradiente
soluciones, feval, gradeval = descenso_gradiente(fobj, gradiente, sol_inicial, step, tolerancia)
```

En lugar de evaluar la norma del gradiente en la solucin dada y validar si es lo suficientemente cercana a cero, se pueden establecer otros criterios de parada para el algoritmo, como:

Tenga en cuenta que su funcin debe retornar 3 listas: la solucin, el valor de la funcin objetivo y el gradiente en cada iteracin; y ser evaluada al utilizarla para resolver el siguiente problema de optimizacin en $\mathbb{R}^4$:

$$ \min f(x) = \frac{1}{2}x^{T}Qx - b^{T}x $$

donde tanto $x$ como $b$ se consideran vectores columna, y:

\begin{equation}
Q=
\begin{bmatrix}

```python
# Definir funcin objetivo - X sera un np.array
def fobj(X):
    Q = np.array([[10, 1, 0, 3], 
                   [1, 6, -6, -3],
                   [0, -6, 12, 6],
                   [3, -3, 6, 9]])
    b = np.array([-8,10,-10,6])
    return (1/2)*np.matmul(np.matmul(np.transpose(X),Q),X)-np.matmul(np.transpose(b),X)

# Definir funcin objetivo - X sera un np.array
def gradiente(X):
    Q = np.array([[10, 1, 0, 3], 
                   [1, 6, -6, -3],
                   [0, -6, 12, 6],
                   [3, -3, 6, 9]])
    b = np.array([-8,10,-10,6])
    return np.matmul(Q,X)-b
```

Recuerde que en ese caso, el gradiente de la funcin se puede calcular como:

$$  \nabla f(x) = Qx - b$$

Implemente su funcin en la siguiente celda de cdigo:

```python
# Algoritmo de descenso de gradiente seleccionando criterio de parada
def descenso_gradiente(fun_obj, gradiente, sol_inicial, step, stop_criteria, stop_value):
    #Ingrese su cdigo ac reemplazando la lnea "raise NotImplementedError"
    # your code here
    soluciones, feval, gradeval = list(), list(), list()
    solucion = sol_inicial
    stop = False
    num_iter = 1
    
    while stop==False:
        gradiente_eval = gradiente(solucion)
        nueva_solucion = solucion - step * gradiente_eval
        solucion_eval = fun_obj(nueva_solucion)
        
        soluciones.append(nueva_solucion)
        feval.append(solucion_eval)
        gradeval.append(gradiente_eval)
        
        print(f"Iteracin {num_iter} - Solucin: {nueva_solucion}, Gradiente: {gradiente_eval}, Funcin objetivo: {solucion_eval:.4f}")
        
        # Criterios de parada
        if stop_criteria == 1:  # Norma del gradiente
            if np.linalg.norm(gradiente_eval) < stop_value:
                stop = True
        elif stop_criteria == 2:  # Cambio en la funcin objetivo
            if num_iter > 1 and abs(feval[-1] - feval[-2]) < stop_value:
                stop = True
        elif stop_criteria == 3:  # Cambio en la solucin
            if num_iter > 1 and np.linalg.norm(soluciones[-1] - soluciones[-2]) < stop_value:
                stop = True
        elif stop_criteria == 4:  # Mximo de iteraciones
            if num_iter >= stop_value:
                stop = True
        
        solucion = nueva_solucion
        num_iter += 1
        
    
    return [soluciones, feval, gradeval]
```
